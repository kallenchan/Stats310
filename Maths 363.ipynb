```python
import numpy as np
import matplotlib.pyplot as plt
# import matplotlib
```

```python
def deltaX1(x,y):
    return np.exp(-y**2-x**4-x*y)*(1+x*(-4*x**3-y))
```

```python
def deltaX2(x,y):
    return np.exp(-y**2-x**4-x*y)*x*(-2*y-x)
```

```python
def Psi(x1,x2):
    return x1*np.exp(-(x2**2+x1**4+x1*x2))
```

```python
def PsiAlongLine(X1,X2):
    return X1*np.exp(-(X2**2+X1**4+X1*X2))
```

```python
#Initialise
X=np.zeros((5,2))
x_0=np.array([0,1.5])
X[0]=x_0
```

```python
k=1
for k in range(1,len(X)):
    x,y=X[k-1]
    X1_k=-deltaX1(x,y)
    X2_k=-deltaX2(x,y)
    X[k]=np.array([X1_k,X2_k])
    k=k+1
```

```python
k=1
for k in range(1,len(X)):
    x,y=X[k-1]
    X1_k=-deltaX1(x,y)
    X2_k=-deltaX2(x,y)
    # X[k]=np.array([X1_k,X2_k])

    #np.linspace(x,x-deltaX1(x,y))
    X1_linspace=np.linspace(x,x-deltaX1(x,y),num=50)
    X2_linspace=np.linspace(y,y-deltaX2(x,y),num=50)
    PsiAlongLineValue=PsiAlongLine(X1_linspace,X2_linspace)

    X[k]=np.array([X1_linspace[np.argmin(PsiAlongLineValue)],
                  X2_linspace[np.argmin(PsiAlongLineValue)]])
    k=k+1
```

```python
x1=np.linspace(-2,1)
x2=np.linspace(-1,2)
X1,X2=np.meshgrid(x1,x2)

Y=X1*np.exp(-(X2**2+X1**4+X1*X2))
import sys
cp=plt.contour(X1,X2,Y,colors='black')

plt.clabel(cp,inline=1,fontsize=10)
cp=plt.contourf(X1,X2,Y,)
plt.xlabel('X1')
plt.ylabel('X2')
plt.scatter(0,1.5,color='red',marker='x')
plt.plot([l[0] for l in X],
           [l[1] for l in X],
           color='red',
        marker='x')
plt.show()
```
